## Support Vector Machines (SVM)

In Sklearn there are three implementations of SVM: LinearSVC, SVC, and NuSVC. They are similar, except that LinearSVC doesn't support kernels. The following code tries LinearSVC and SVC (with a few kernel options) on the Iris dataset. The decision boundaries can be easily affected by different kernels and C values.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import svm, datasets




def svm_tester(C,h,X,y,Xtest=None,ytest=None):

    svc = svm.SVC(kernel='linear', C=C).fit(X, y)
    rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, y)
    poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X, y)
    lin_svc = svm.LinearSVC(C=C).fit(X, y)

    
    
    # scores from various kernels
    if Xtest is None:
        svc_score = svc.score(X,y)
        rbf_score = rbf_svc.score(X,y)
        poly_score = poly_svc.score(X,y)
        linear_score = lin_svc.score(X,y)
    else:
        svc_score = svc.score(Xtest,ytest)
        rbf_score = rbf_svc.score(Xtest,ytest)
        poly_score = poly_svc.score(Xtest,ytest)
        linear_score = lin_svc.score(Xtest,ytest)

    """
    print 'C value: ', C
    # check on performance
    print 'SVC:', svc_score
    print 'RBF_SVC:', rbf_score
    print 'Poly_SVC:', poly_score
    print 'Linear_SVC:', linear_score
    print "\n"
    """
    
    
       
    # create a mesh to plot in
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # title for the plots
    titles = ['SVC with linear kernel',
              'LinearSVC (linear kernel)',
              'SVC with RBF kernel',
              'SVC with polynomial (degree 3) kernel']

    for i, clf in enumerate((svc, lin_svc, rbf_svc, poly_svc)):
        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, m_max]x[y_min, y_max].
        plt.subplot(2, 2, i + 1)
        plt.subplots_adjust(wspace=0.4, hspace=0.4)

        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

        # Plot also the training points
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
        plt.xlabel('Sepal length')
        plt.ylabel('Sepal width')
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
        plt.xticks(())
        plt.yticks(())
        plt.title(titles[i])
        
    return np.array([svc_score,rbf_score,poly_score,linear_score],dtype='f')
    #print output.shape
    #print output
        
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features. 
y = iris.target

h = .02  # step size in the mesh

# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 0.1  # SVM regularization parameter        



output = svm_tester(C,h,X,y)


for i in range (1,50):
    C+=0.1
    output = np.c_[output,svm_tester(C,h,X,y)]
  


#Feels like a very convoluted way to do this, is there a simpler way to extract these data?
output = np.swapaxes(output,1,0)


labels = [
    "SVC, linear kernel",
    "SVC, RBF kernel:",
    "SVC, 3rd degree polynomial kernel",
    "Linear_SVC"
]

points = [
    '-ro',
    '-g+',
    '-bo',
    '-c+'
]



plt.figure(0)
for i in range (0,4):
    plt.plot(output[:,i],points[i],label=labels[i])   
    

#plt.xlim(0, 1) 
#plt.xticks(np.arange(0.0,10,1))
plt.ylabel('score')
plt.xlabel('Change in C value, interval = 0.1')
plt.legend(loc='best')

plt.show()

```

## Combat class imbalance

The SVC package allows us to specify class weights so as to cope with imbalance in training data. The following code generates two classes of data: 1000 random points around (0,0) and 100 random points around (2,2). The hyperplane with or without applying class weights is shown, along with the fitting score of the corresponding SVM.

```python
# we create 40 separable points
rng = np.random.RandomState(0)
n_samples_1 = 1000
n_samples_2 = 100
X = np.r_[1.5 * rng.randn(n_samples_1, 2),
          0.5 * rng.randn(n_samples_2, 2) + [2, 2]]
y = [0] * (n_samples_1) + [1] * (n_samples_2)



# fit the model and get the separating hyperplane
clf = svm.SVC(kernel='linear', C=1.0)
clf.fit(X, y)
print 'Score without weighting:',clf.score(X,y)

w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - clf.intercept_[0] / w[1]

# get the separating hyperplane using weighted classes
wclf = svm.SVC(kernel='linear', class_weight={1: 10})
wclf.fit(X, y)
print 'Score after weighting:',wclf.score(X,y)

ww = wclf.coef_[0]
wa = -ww[0] / ww[1]
wyy = wa * xx - wclf.intercept_[0] / ww[1]



# plot separating hyperplanes and samples
h0 = plt.plot(xx, yy, 'k-', label='no weights')
h1 = plt.plot(xx, wyy, 'k--', label='with weights')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.legend(loc='best')

plt.axis('tight')
```

**Question:** Do you note any difference on the fitting score? Can you explain why?

**Answer:**

Yes, the weighted score is lower than unweighted. This is because of the increase in false positives towards the smaller class. As with weights, we get lowered false negative, but because the data is not linearly seperatable, the number of false positives would also increase.

## Novelty detection

SVM can be extended to one-class classifiers that make effective novelty detectors. The following code generates two clouds of "normal" data and then some outliers. SVM does a good job in learning the boundaries of normal data and detect potential outliers. You can read more about the <a href="http://scikit-learn.org/stable/modules/outlier_detection.html#outlier-detection">novelty detection</a> and check out <a href="http://scikit-learn.org/stable/modules/generated/sklearn.svm.OneClassSVM.html#sklearn.svm.OneClassSVM">OneClassSVM</a>. 

```python
import matplotlib.font_manager

xx, yy = np.meshgrid(np.linspace(-5, 5, 500), np.linspace(-5, 5, 500))
# Generate train data
X = 0.3 * np.random.randn(100, 2)
X_train = np.r_[X + 2, X - 2]
# Generate some regular novel observations
X = 0.3 * np.random.randn(20, 2)
X_test = np.r_[X + 2, X - 2]
# Generate some abnormal novel observations
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))

# fit the model
clf = svm.OneClassSVM(nu=0.1, kernel="rbf", gamma=0.1)
clf.fit(X_train)
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)
y_pred_outliers = clf.predict(X_outliers)
n_error_train = y_pred_train[y_pred_train == -1].size
n_error_test = y_pred_test[y_pred_test == -1].size
n_error_outliers = y_pred_outliers[y_pred_outliers == 1].size

# plot the line, the points, and the nearest vectors to the plane
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.title("Novelty Detection")
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap=plt.cm.Blues_r)
a = plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='red')
plt.contourf(xx, yy, Z, levels=[0, Z.max()], colors='orange')

b1 = plt.scatter(X_train[:, 0], X_train[:, 1], c='white')
b2 = plt.scatter(X_test[:, 0], X_test[:, 1], c='green')
c = plt.scatter(X_outliers[:, 0], X_outliers[:, 1], c='red')
plt.axis('tight')
plt.xlim((-5, 5))
plt.ylim((-5, 5))
plt.legend([a.collections[0], b1, b2, c],
           ["learned frontier", "training observations",
            "new regular observations", "new abnormal observations"],
           loc="upper left",
           prop=matplotlib.font_manager.FontProperties(size=11))
plt.xlabel(
    "error train: %d/200 ; errors novel regular: %d/40 ; "
    "errors novel abnormal: %d/40"
    % (n_error_train, n_error_test, n_error_outliers))
```

```python
# fit the model

nu_value = 0.1

n_error_test_list = []
n_error_outliers_list = []


for i in range (0,10):
    clf = svm.OneClassSVM(nu=nu_value, kernel="rbf", gamma=0.1)
    clf.fit(X_train)
    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)
    y_pred_outliers = clf.predict(X_outliers)
    n_error_train = y_pred_train[y_pred_train == -1].size

    n_error_test = y_pred_test[y_pred_test == -1].size
    n_error_test_list.append(n_error_test)
    n_error_outliers = y_pred_outliers[y_pred_outliers == 1].size
    n_error_outliers_list.append(n_error_outliers)
    nu_value += 0.1

    
    
plt.figure(0)
plt.plot( n_error_test_list, '-ro', label='False positives')
plt.plot( n_error_outliers_list, '-g+', label='False negatives')


plt.ylabel('score')
plt.xlabel('Change in nu value, interval = 0.1')
plt.legend(loc='best')

plt.show()

```

- As nu value increases, the learned frontier decreases. The number of false positives go up, false negatives generally remain in the same value, decreasing slightly towards1 as nu goes down.

- Changing the gamma value produces no significant difference.
